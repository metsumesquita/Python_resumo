```python
def fatorial(numero):
  if numero==0 or numero==1:
    return 1
  resultado=1
  for i in range(2,numero+1):
    resultado*=i
  return resultado
```

```python
print(fatorial(5))
```

```python
#mean

beans_values=[10,8,12,15,17,6,9]

mean = sum(beans_values)/len(beans_values)

print("A média é {r:.2f}".format(r=mean))
```

```python
#desvio padrao
soma = 0

for valor_atual in beans_values:
    diferença = valor_atual - mean
    soma += diferença ** 2

desvio_padrao = math.sqrt(soma / len(beans_values))

print("O desvio padrão populacional é {r:.2f}".format(r=desvio_padrao))
```

#series
Uma Series do Pandas é construída sobre um array NumPy, mas adiciona funcionalidades extras, principalmente um sistema de indexação mais flexível.
Indexação Flexível: Diferente dos arrays NumPy que usam índices numéricos padrão (0, 1, 2...), as Series permitem o uso de rótulos personalizados (strings, datas, etc.) como índices.
Heterogeneidade Potencial: Embora geralmente se usem dados do mesmo tipo, as Series são um pouco mais flexíveis na prática em relação aos tipos de dados.
Análise de Dados: São ideais para representar colunas individuais de uma tabela (DataFrame) e são a base para a análise e manipulação de dados no ecossistema Pandas
-
Series são arrays unidimensionais capazes de armazenar qualquer tipo de dado - basicamente vetores.
 Cada elemento das Series tem um índice associado. Elas podem ser usada para representar séries temporais, sequências de valores, ou listas de elementos unidimensionais. A criação delas é feita a partir de listas, dicionários, ou arrays do NumPy.
```python
df_heath = df_heath.rename(columns={"age": "Age","bmi":"BMI",'sex':'Sex',"children":"Children","smoker":"Smoker","region":"Region","charges":"Charges"})
```

```python
#iniciando do primeiro ate ultimo
df_heath['Id_cliente'] = pd.Series(range(1,len(df_heath)+1))
```

```python
# 1. Define conditions (7 conditions)
cond_list = [
    (df_heath['Age'] >= 18) & (df_heath['BMI'] < 18.5),                # Underweight
    (df_heath['Age'] >= 18) & (df_heath['BMI'] >= 18.5) & (df_heath['BMI'] <= 24.9), # Healthy
    (df_heath['Age'] >= 18) & (df_heath['BMI'] >= 25.0) & (df_heath['BMI'] <= 29.9), # Overweight
    (df_heath['Age'] >= 18) & (df_heath['BMI'] >= 30.0) & (df_heath['BMI'] <= 34.9), # Class 1
    (df_heath['Age'] >= 18) & (df_heath['BMI'] >= 35.0) & (df_heath['BMI'] <= 39.9), # Class 2
    (df_heath['Age'] >= 18) & (df_heath['BMI'] >= 40.0)                # Severe
]

# 2. Define choices (7 choices - MUST match condition list length)
choice_list = ['Underweight', 'Healthy Weight', 'Overweight',
               'Class 1 Obesity', 'Class 2 Obesity', 'Severe Obesity']

# 3. Apply conditions
df_heath['Risk Categories'] = np.select(cond_list, choice_list, default='Other/Adult')

print(df_heath)
```

```python
# Reordering columns using loc
df_heath = df_heath.loc[:, ["Id_cliente","Sex", "Age","Smoker","BMI","Risk Categories","Children","Region","Charges"]]
print(df_heath.head())
```
